```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data full clean.csv")
display(df.head())
print("df shape: ", df.shape)
print("column names:", df.columns.tolist())
```

```python
def change(df, cols):
    lis = []
    for i in df[cols]:
        i = str(i).replace(".",",")
        lis.append(i)
    df[cols] = lis
    return df
```

```python
cols = ["rent","rooms","size"]
for i in cols:
    df = change(df,i)
```
